```python
from joblib import Parallel, delayed
df = pd.read_csv("covid-19-bogota.csv",sep=";",encoding="latin-1")
df.dropna(inplace=True)
df["Día 2020 síntomas"] = (pd.to_datetime(df["Fecha de inicio de síntomas"]) - pd.to_datetime("1/1/2020")).astype(int)/(10**9*60*60*24)
df["Día 2020 diagnóstico"] = (pd.to_datetime(df["Fecha de diagnóstico"]) - pd.to_datetime("1/1/2020")).astype(int)/(10**9*60*60*24)
```

```python
df_sub = df[['Edad','Día 2020 síntomas','Día 2020 diagnóstico']]
df_sub
```

```python
def maximo(fila):
    if not np.isnan(fila).any():
        names = ['Edad','Día 2020 síntomas','Día 2020 diagnóstico']
        return names[fila.index(max(fila))]
    else:
        return "None"
        
```

```python
results = []
for fila in df_sub.values.tolist():
    results.append(maximo(fila))
```

```python
results
```

```python
result = Parallel(n_jobs = 2, backend = "multiprocessing")(
map(delayed(maximo), df_sub.values.tolist()))
```

```python
result
```

### Clusters (K-Means)

```python
from sklearn.cluster import KMeans
df = pd.read_csv("covid-19-bogota.csv",sep=";",encoding="latin-1", nrows=1e4)
df.dropna(inplace=True)
df["fecha_sintomas"] = pd.to_datetime(df["Fecha de inicio de síntomas"],format='%d/%m/%Y')
df["fecha_diagnostico"] = pd.to_datetime(df["Fecha de diagnóstico"],format='%d/%m/%Y')
df["dia_sintomas"] = (df["fecha_sintomas"] - pd.to_datetime("1/1/2020",format='%d/%m/%Y')).astype(int)/(10**9*60*60*24)
df["dia_diagnostico"] = (df["fecha_diagnostico"] - pd.to_datetime("1/1/2020",format='%d/%m/%Y')).astype(int)/(10**9*60*60*24)
df["espera"] = df["dia_diagnostico"] - df["dia_sintomas"]
df["mayo"] = df["dia_diagnostico"] > (31+29+31+30)
newdf = df[["Edad","dia_sintomas","espera"]]
kmeans = KMeans(n_clusters=4, random_state=0,n_jobs = 1).fit(newdf)
kmeans.labels_
```

#### Variables categóricas

```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("covid-19-bogota.csv",sep=";",encoding="latin-1").dropna()
df["fecha_sintomas"] = pd.to_datetime(df["Fecha de inicio de síntomas"],format='%d/%m/%Y')
df['ano_sintomas'] = pd.DatetimeIndex(df["fecha_sintomas"]).year
df['mes_sintomas'] = pd.DatetimeIndex(df["fecha_sintomas"]).month
df['dia_sintomas'] = pd.DatetimeIndex(df["fecha_sintomas"]).day
df["fecha_diagnostico"] = pd.to_datetime(df["Fecha de diagnóstico"],format='%d/%m/%Y')
df['ano_diagnostico'] = pd.DatetimeIndex(df["fecha_diagnostico"]).year
df['mes_diagnostico'] = pd.DatetimeIndex(df["fecha_diagnostico"]).month
df['dia_diagnostico'] = pd.DatetimeIndex(df["fecha_diagnostico"]).day
df['duracion'] = df["fecha_diagnostico"] - df["fecha_sintomas"]
df['duracion'] = df['duracion'].dt.days
```

```python
df_prueba = df[(df['ano_sintomas'] >= 2020) & (df['ano_diagnostico'] >= 2020)]
Y = df_prueba[['duracion']]
X = df_prueba[['Edad']]
df_prueba['mes_diagnostico_str'] = df_prueba['mes_diagnostico'].apply(str)

dummies = pd.get_dummies(data=df_prueba[['mes_diagnostico_str','Localidad de residencia','Sexo']])
X = dummies.add(X,fill_value=0)
X.columns
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X,Y)
```
